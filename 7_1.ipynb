## 7.1. Grafici in numpy e matplotlib

Come sappiamo fare un grafico di funzione in generale è complesso e non ci sono regole universali, se non disegnare il grafico per punti e intuirne l'andamento. Certo, lo studio delle derivare e dei limiti ai bordi del dominio possono aiutare. lo vedremo più avanti.

Il vantaggio di avere un calcolatore è poter valutare una funzione in un ampio numero di punti del dominio.

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x)**2*x**2
```

```python
x=np.linspace(-10,10,50)#50 punti equispaziati nell'intervallo [-10, 10]
plt.plot(x,f(x))
```

Un altro modo di definire una funzione da valutare in una griglia di punti è usare la funzione inline "lambda"

```python
f=lambda x: np.sin(x**2)*x**2
```

```python
x=np.linspace(-10,10,1000)
plt.plot(x,f(x))
```

```python
f= lambda x: np.sin(x**-2)**2*x**2
x=np.linspace(-1,1,1000)
plt.plot(x,f(x))
```

```python
#Vicino ai punti di discontinuità plt.plot unirà i punti vicini, spezzando la linea per NaN o Inf
x=np.linspace(-10,10)
plt.plot(x,1/x)
```

```python
#per evitare questo effetto, disegnamo la funzione includendo il punto di discontinuità
x=np.linspace(-1,1,501)
plt.plot(x,1/x); plt.grid()
```
